```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = plt.imread('pink_lake.png')
plt.imshow(img)
```

```python
img_add = np.clip(img + 0.25, 0, 1)
plt.imsave('img_add.png', img_add)
plt.imshow(img_add)
```

```python
def extract_channel(img, dim):
    new = np.zeros_like(img)
    new[:,:,dim] = img[:,:,dim]
    return new
```

```python
img_chan_0 = extract_channel(img, 0)
img_chan_1 = extract_channel(img, 1)
img_chan_2 = extract_channel(img, 2)

plt.imsave('img_chan_0.png', img_chan_0)
plt.imsave('img_chan_1.png', img_chan_1)
plt.imsave('img_chan_2.png', img_chan_2)
```

```python
img_gray = img_chan_0[:,:,0] * 0.299 + img_chan_1[:,:,1] * 0.587 + img_chan_2[:,:,2] * 0.114
img_gray = np.tile(img_gray[:,:,np.newaxis], (1,1,3))
plt.imsave('img_gray.png', img_gray)
```

```python
img_crop = img_gray[:img_gray.shape[0]//2,:,:]
plt.imsave('img_crop.png', img_crop)
```

```python
img_flip_vert = np.flip(img, axis=0)
plt.imsave('img_flip_vert.png', img_flip_vert)
```
